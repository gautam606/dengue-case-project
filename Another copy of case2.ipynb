```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.calibration import calibration_curve

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
features = pd.read_csv("DengAI_Predicting_Disease_Spread_-_Training_Data_Features.csv")
labels = pd.read_csv("DengAI_Predicting_Disease_Spread_-_Training_Data_Labels.csv")

data = features.merge(labels, on=["city","year","weekofyear"])

print(data.head())
```

```python
sj = data[data["city"]=="sj"]
```

```python
annual = sj.groupby("year")["total_cases"].sum().reset_index()

print(annual.head())
```

```python
  annual["growth_rate_%"] = annual["total_cases"].pct_change()*100
annual["growth_rate_%"].fillna(0, inplace=True)

print(annual)
```

```python
peak = annual.loc[annual["total_cases"].idxmax()]

print("Peak Year:", peak["year"])
print("Maximum Cases:", peak["total_cases"])
```

```python
cases = annual["total_cases"].values.reshape(-1,1)

scaler = MinMaxScaler()
cases_scaled = scaler.fit_transform(cases)

X = []
y = []

for i in range(3, len(cases_scaled)):
    X.append(cases_scaled[i-3:i])
    y.append(cases_scaled[i])

X = np.array(X)
y = np.array(y)
```

```python
  model = Sequential()

model.add(LSTM(50, activation="relu", input_shape=(X.shape[1],1)))
model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")

model.summary()
```

```python
model.fit(X, y, epochs=200, verbose=1)
```

```python
train_predictions = model.predict(X)

train_predictions = scaler.inverse_transform(train_predictions)
actual = scaler.inverse_transform(y)
```

```python
plt.figure()
plt.plot(actual, label="Actual")
plt.plot(train_predictions, label="Predicted")
plt.title("Actual vs Predicted Dengue Cases")
plt.legend()
plt.show()
```

```python
future_predictions = []

current_input = cases_scaled[-3:]

for i in range(5):

    pred = model.predict(current_input.reshape(1,3,1))

    future_predictions.append(pred[0][0])

    current_input = np.append(current_input[1:],pred,axis=0)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1,1))

future_years = np.arange(annual["year"].max()+1, annual["year"].max()+6)
```

```python
plt.figure()

plt.plot(annual["year"],annual["total_cases"],label="Historical")

plt.plot(future_years,future_predictions,label="Forecast")

plt.title("5 Year Dengue Forecast")

plt.legend()

plt.show()
```

```python
threshold = np.mean(actual)

actual_class = (actual > threshold).astype(int)

pred_class = (train_predictions > threshold).astype(int)
```

```python
accuracy = accuracy_score(actual_class,pred_class)

print("Accuracy:",accuracy)
```

```python
precision = precision_score(actual_class,pred_class)

print("Precision:",precision)
```

```python
f1 = f1_score(actual_class,pred_class)

print("F1 Score:",f1)
```

```python
cm = confusion_matrix(actual_class,pred_class)

sns.heatmap(cm,annot=True,fmt="d")

plt.title("Confusion Matrix")

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.show()
```

```python

prob_true,prob_pred = calibration_curve(actual_class,pred_class,n_bins=5)

plt.plot(prob_pred,prob_true,marker="o")

plt.plot([0,1],[0,1])

plt.title("Reliability Diagram")

plt.xlabel("Predicted Probability")

plt.ylabel("True Probability")

plt.show()
```

```python
threshold = sj["total_cases"].quantile(0.75)

sj["outbreak"] = (sj["total_cases"] > threshold).astype(int)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate(name, y_true, y_pred):
    print(f"\n{name}")
    print("Accuracy:", round(accuracy_score(y_true, y_pred),2))
    print("Precision:", round(precision_score(y_true, y_pred),2))
    print("Recall:", round(recall_score(y_true, y_pred),2))
    print("F1 Score:", round(f1_score(y_true, y_pred),2))

evaluate("Linear Regression", y_test, y_pred_lr)
evaluate("Random Forest", y_test, y_pred_rf)
evaluate("XGBoost", y_test, y_pred_xgb)
```
